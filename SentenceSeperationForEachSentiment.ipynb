```python
import pandas as pd
import csv


df = pd.read_csv("SigmaLaw-ABSA.csv")
file_name = 'SigmaLaw-ABSA-PartyLabel'


count=0
# This code to process the input data with out any party Label
# import pandas as pd
# import csv
#
# df = pd.read_csv("SigmaLaw-ABSA.csv")
# file_name = 'Part1'
#
# count=0
# with open('%s.csv' % (file_name), 'w+') as file:
#     w = csv.writer(file)
#     w.writerow(['Sentence', 'Party', 'Sentiment'])
#     for index, row in df.iterrows():
#         count=count+1
#         print(count)
#         sentence = row["Sentence"]
#         sentimentList = row["Sentiment"]
#         sentimentArray = sentimentList.split('],')
#         aspectList = row["Party"]
#         aspectArray = aspectList.split('],')
#
#         for i in range(len(aspectArray)):
#             aspect=None
#             sentiment=None
#             aspect = aspectArray[i].replace('[', '').replace(']', '')
#             try:
#                 sentiment = sentimentArray[i].replace('[', '').replace(']', '')
#             except Exception as e:
#                 print(str(sentimentArray)+":"+str(i) + " : "+aspect)
#
#             if not sentiment == "":
#                 if ',' in sentiment:
#                     sentiment = sentiment.split(",")[0]
#                 else:
#                     sentiment = sentiment
#
#             if not aspect == "":
#                 if ',' in aspect:
#                     aspect = aspect.split(",")[0]
#                     if "_" in aspect:
#                         aspect = aspect.split("_")[0]
#                 else:
#                     aspect = aspect
#                     if "_" in aspect:
#                         aspect = aspect.split("_")[0]
#             if sentence and (aspect or sentiment):
#                 w.writerow([sentence, aspect, sentiment])
#                 print(sentence," : " ,aspect, " : ", sentiment)



import pandas as pd
import csv

df = pd.read_csv("SigmaLaw-ABSA-PartyLabel.csv")
file_name = 'Part1'

count=0
with open('%s.csv' % (file_name), 'w+') as file:
    w = csv.writer(file)
    w.writerow(['Sentence', 'Party', 'PartyLabel','Sentiment'])
    for index, row in df.iterrows():
        count=count+1
        print(count)
        sentence = row["Sentence"]
        sentimentList = row["Sentiment"]
        sentimentArray = sentimentList.split('],')

        partyLabelList = row["PartyLabel"]
        partyLabelArray = partyLabelList.split('],')

        aspectList = row["Party"]
        aspectArray = aspectList.split('],')

        for i in range(len(aspectArray)):
            aspect=None
            sentiment=None
            aspect = aspectArray[i].replace('[', '').replace(']', '')
            PartyLabel = partyLabelArray[i].replace('[', '').replace(']', '')
            try:
                sentiment = sentimentArray[i].replace('[', '').replace(']', '')
            except Exception as e:
                print(str(sentimentArray)+":"+str(i) + " : "+aspect)

            if not len(sentiment) == 0:
                if ',' in sentiment:
                    sentiment = sentiment.split(",")[0]
                else:
                    sentiment = sentiment

            if not len(PartyLabel) == 0:
                if ',' in PartyLabel:
                    PartyLabel = PartyLabel.split(",")[0]
                else:
                    PartyLabel = PartyLabel

            if not len(aspect) == 0:
                if ',' in aspect:
                    aspect = aspect.split(",")[0]
                    if "_" in aspect:
                        aspect = aspect.split("_")[0]
                else:
                    aspect = aspect
                    if "_" in aspect:
                        aspect = aspect.split("_")[0]
            if sentence and (aspect or sentiment):
                w.writerow([sentence, aspect, PartyLabel, sentiment])
                print(sentence," : " ,aspect, " : ",PartyLabel, " : ",sentiment)












partyLabelList=[]
lst1=[]
list2=[]
with open('%s.csv' % (file_name), 'w+') as file:
    w = csv.writer(file)
    w.writerow(['Sentence', 'Party', 'PartyLabel','Sentiment','Overall Sentiment'])
    for index, row in df.iterrows():
        count = count+1

        party = row["Party"]
        #party = "[[lee,[]]"
        aspectArray = party.split('],[')
        try:

            partyLabelList = []
            list1 = []
            list2 = []
            if ',' in aspectArray[0]:
                list1 = [1] * (aspectArray[0].count(',') + 1)
            else:
                str1 = aspectArray[0].replace('[', "").replace(']', "")
                if not len(str1) == 0:
                    list1 = [1]
            if ',' in aspectArray[1]:
                list2 = [0] * (aspectArray[1].count(',') + 1)
            else:
                str2 = aspectArray[1].replace('[', "").replace(']', "")
                if not len(str2) == 0:
                    list2 = [0]
        except Exception as ex:
            print(party+":"+str(row))
            print('====\n')



        partyLabelList.append(list1)
        partyLabelList.append(list2)

        w.writerow([row["Sentence"], row["Party"],"".join(str(partyLabelList)), row["Sentiment"],row["Overall Sentiment"]])




```
